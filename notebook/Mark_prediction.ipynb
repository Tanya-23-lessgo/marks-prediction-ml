```python
# Install dependencies as needed:
# pip install kagglehub[pandas-datasets]
import kagglehub
from kagglehub import KaggleDatasetAdapter

path=kagglehub.dataset_download("whenamancodes/student-performance")
print("Path to dataset files:", path)
```

```python
import os
path="/kaggle/input/student-performance"
os.listdir(path)
```

```python
import pandas as pd
df=pd.read_excel(f"{path}/Maths.csv")
```

```python
df.head()
```

```python
df.info()
```

```python
to_drop = [
    "school","famsize","Pstatus","Mjob","Fjob","reason","guardian",
    "traveltime","famsup","paid","activities","nursery","higher",
    "internet","romantic",
    "sex","address"
]

df2 = df.drop(columns=to_drop)
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
```

```python
for col in df2.columns:
  if df2[col].dtype=="object":
    df2[col]=le.fit_transform(df2[col])
```

```python
df.isnull().sum()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error,r2_score
```

```python
y=df2["G3"]
x=df2.drop("G3",axis=1)
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
model=LinearRegression()
model.fit(x_train,y_train)
```

```python
y_pred=model.predict(x_test)
```

```python
print("Mean absolute error:",mean_absolute_error(y_test,y_pred))
```

```python
print("R2 Error:",r2_score(y_test,y_pred))

```

```python
sample=x_test.iloc[0]
predicted_mark=model.predict([sample])
print(f"\nPredicted mark:{predicted_mark}")
print("Actual mark:", y_test.iloc[0])
```
